```python
import matplotlib.pyplot as plt

validCodewords = {
    '0000': '11110', '0001': '01001', '0010': '10100', '0011': '10101',
    '0100': '01010', '0101': '01011', '0110': '01110', '0111': '01111',
    '1000': '10010', '1001': '10011', '1010': '10110', '1011': '10111',
    '1100': '11010', '1101': '11011', '1110': '11100', '1111': '11101'
}

dataWord = "10101010"

codeWord = ""
for i in range(0, len(dataWord), 4):
    bloco_4b = dataWord[i : i + 4]
    bloco_5b = validCodewords[bloco_4b]
    codeWord += bloco_5b

print(f"Sinal de Entrada: {dataWord}")
print(f"Saída (4B/5B): {codeWord}")
```

```python
NRZI = []
current_voltage = -1

for bit in codeWord:
    if bit == '1':
        current_voltage = current_voltage * -1

    NRZI.append(current_voltage)
plt.figure(figsize=(12, 3))

# Desenhando o gráfico em formato de degrau
plt.step(range(len(NRZI)), NRZI, where='post', linewidth=2, color='blue')

# Ajustes estéticos
plt.ylim(-1.5, 1.5)
plt.yticks([-1, 0, 1], ['-V', '0', '+V'])
plt.xticks([i for i in range(len(codeWord))], list(codeWord))
plt.axhline(0, color='black', linewidth=1) # Eixo central
plt.ylabel("Nível de Sinal")
plt.xlabel("Bits Codificados (4B/5B)")
plt.title(f"Sinal 4B/5B com Transmissão NRZ-I")
plt.grid(True, axis='x', linestyle='--', alpha=0.7)

plt.show()
```
