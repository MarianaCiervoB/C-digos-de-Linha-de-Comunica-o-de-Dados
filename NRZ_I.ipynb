```python
import matplotlib.pyplot as plt

NRZ = [1, 0, 1, 0, 1, 0, 1, 0]
NRZI = []
current_voltage = -1

for bit in NRZ:
    if bit == 1:
        current_voltage = current_voltage * -1

    NRZI.append(current_voltage)

print(f"Entrada: {NRZ}")
print(f"Saída NRZ-I: {NRZI}")
```

```python
plt.figure(figsize=(10, 3))
plt.step(range(len(NRZI)), NRZI, where='post', linewidth=2, color='red')

# ajustes
plt.ylim(-1.5, 1.5)
plt.yticks([-1, 0, 1], ['-V', '0', '+V'])
plt.xticks([i for i in range(len(NRZ))], NRZ)
plt.axhline(color='black', linewidth=1) # Linha do meio
plt.ylabel("Nível de Sinal")
plt.xlabel("Sequência de Bits")
plt.title("NRZ-I (Non-Return-to-Zero Inverted)")
plt.grid(True, axis='x', linestyle='--', alpha=0.7) # linhas verticais para os intervalos de bit

plt.show()
```
